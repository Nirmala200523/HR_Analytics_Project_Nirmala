```python
import pandas as pd

df = pd.read_excel("../Excel/HR_Attrition_Cleaned_Final.xlsx")

df.head()
```

```python
df = df.drop(columns=['Unnamed: 36', 'Unnamed: 37', 'Total Employees', 1470], errors='ignore')

df.info()
```

# Cell 1: Attrition Count (Bar Chart)

```python
df['Attrition'].value_counts().plot(kind='bar', figsize=(5,4))

plt.title("Employee Attrition")
plt.xlabel("Attrition")
plt.ylabel("Count")

plt.show()
```

# Cell 2: Department-wise Employees

```python
df['Department'].value_counts().plot(kind='bar', figsize=(6,4))

plt.title("Employees by Department")
plt.xlabel("Department")
plt.ylabel("Count")

plt.show()
```

# Cell 3: Age Distribution

```python
df['Age'].plot(kind='hist', bins=10, figsize=(6,4))

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Employees")

plt.show()
```

# Cell 4: Job Satisfaction

```python
df['JobSatisfaction'].value_counts().sort_index().plot(kind='bar')

plt.title("Job Satisfaction")
plt.xlabel("Satisfaction Level")
plt.ylabel("Employees")

plt.show()
```

# Cell 5: Attrition by Overtime

```python
df.groupby('OverTime')['Attrition'].value_counts().unstack().plot(kind='bar', figsize=(6,4))

plt.title("Attrition by Overtime")
plt.xlabel("OverTime")
plt.ylabel("Employees")

plt.show()
```

# Cell 6: Monthly Income Distribution

```python
df['MonthlyIncome'].plot(kind='hist', bins=20, figsize=(6,4))

plt.title("Monthly Income Distribution")
plt.xlabel("Monthly Income")
plt.ylabel("Employees")

plt.show()
```

# Cell 7: Correlation Heatmap

```python
import matplotlib.pyplot as plt

corr = df.select_dtypes(include=['int64','float64']).corr()

plt.figure(figsize=(12,10))
plt.imshow(corr, cmap='coolwarm', aspect='auto')
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90, fontsize=8)
plt.yticks(range(len(corr.columns)), corr.columns, fontsize=8)

plt.title("Correlation Heatmap")
plt.tight_layout()

plt.show()
```
